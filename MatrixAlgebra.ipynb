```python
import numpy as np
```

```python
#Función para suma de matrices
def Sum(a,b):
    matrix1=np.zeros((a,b))
    matrix2=np.zeros((a,b))
    matrix3=np.zeros((a,b))
    #Rellenando primera matriz
    print("\nPrimera matriz: ")
    for i in range(a):
        for j in range(b):
            x=float(input("Ingresa un valor: "))
            matrix1[i][j]=x
    #Rellenando segunda matriz
    print("\nSegunda matriz: ")
    for i in range(a):
        for j in range(b):
            x=float(input("Ingresa un valor: "))
            matrix2[i][j]=x
    #Suma de matrices
    print("\nMatriz resultante: ")
    for i in range(a):
        for j in range(b):
            matrix3[i][j]=matrix1[i][j]+matrix2[i][j]
    return matrix3
#Función para resta de matrices
def Res(a,b):
    matrix1=np.zeros((a,b))
    matrix2=np.zeros((a,b))
    matrix3=np.zeros((a,b))
    #Rellenando primera matriz
    print("\nPrimera matriz: ")
    for i in range(a):
        for j in range(b):
            x=float(input("Ingresa un valor: "))
            matrix1[i][j]=x
    #Rellenando segunda matriz
    print("\nSegunda matriz: ")
    for i in range(a):
        for j in range(b):
            x=float(input("Ingresa un valor: "))
            matrix2[i][j]=x
    #Suma de matrices
    print("\nMatriz resultante: ")
    for i in range(a):
        for j in range(b):
            matrix3[i][j]=matrix1[i][j]-matrix2[i][j]
    return matrix3
#Función para multiplicación de matrices
def Mult(a,b):
    matrix1=np.zeros((a,b))
    matrix2=np.zeros((a,b))
    matrix3=np.zeros((a,b))
    #Rellenando primera matriz
    print("\nPrimera matriz: ")
    for i in range(a):
        for j in range(b):
            x=float(input("Ingresa un valor: "))
            matrix1[i][j]=x
    #Rellenando segunda matriz
    print("\nSegunda matriz: ")
    for i in range(a):
        for j in range(b):
            x=float(input("Ingresa un valor: "))
            matrix2[i][j]=x
    #Multiplicación de matrices
    print("\nMatriz resultante: ")
    for i in range(a):
        for j in range(b):
            for k in range(a):
                 matrix3[i][j]=matrix3[i][j]+matrix1[i][k]*matrix2[k][j]
    return matrix3
#Determinante matrix 2x2
def Detr(a,b):
    matrix1=np.zeros((a,b))
    f1=1
    f2=1
    determinante=0
    for i in range(a):
        for j in range(b):
            x=float(input("Ingresa un valor: "))
            matrix1[i][j]=x
    for i in range(a):
        for j in range(b):
            if(i==j):
                f1=f1*matrix1[i][j]
            else:
                f2=f2*matrix1[i][j]
    determinante=f1-f2
    return determinante
```

```python
#Here you have a code for Matrix Algebra
a=int(input("Ingrese el número de columnas: "))
b=int(input("Ingrese el número de renglones: "))
print(Sum(a,b))
print(Res(a,b))
print(Mult(a,b))
print(Detr(a,b))
```
